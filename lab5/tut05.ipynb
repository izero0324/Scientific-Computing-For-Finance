```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
from numpy import pi, sin, cos

x_domain = np.arange(0, 2*pi, 5e-2)
y_array = cos(x_domain) + sin(x_domain)

plt.plot(x_domain, y_array)
```

```python
from numpy import exp
x = np.arange(0,1, 1e-4)
y = exp(x)

plt.plot(x, y)
```

```python
y2 = cos(x) + sin(x)
plt.plot(x, y, x,y2)
```

```python
plt.plot(np.matrix([y,y2]).T)
```
