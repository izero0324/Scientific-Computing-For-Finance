# Exercise 1
1. Create a sequence n = 15 normally distributed


```python
data = np.random.normal(size=15)
```

2. Create a time-series from data using the pandas and call it time_series

```python
time_series = pd.Series(data)
```

3. Label time_series from the date 20/9/2017

```python
dates = pd.date_range('20/9/2017', periods=15)
time_series = pd.Series(data, index=dates)
print(time_series)
```

4. Print the value corresponding to the date 23/9/2017

```python
print(time_series['23/9/2017'])
```

5. plot time_series using pandas

```python
time_series.plot()
```

6. plot the cumulative sum of time_series using pandas

```python
time_series.cumsum().plot()
```

# Exercies 3
1. Retrieve stock price daily data for 'Goldman Sachs', (e.g. indicator 'GS'), between dates 1/1/2010 and 27/1/2013) from Yahoo Finance using pandas.

```python
import yfinance as yfin
from pandas_datareader import data
yfin.pdr_override()

gsframe = data.get_data_yahoo('GS', start='2010-01-01', end='2014-01-27')

print(gsframe.head())
```

```python
print(gsframe.tail())
```

2. Plot the 'High' and 'Low' prices in one figure.

```python
high_low = gsframe[['High', 'Low']]
high_low.plot()
```

3. Plot the empirical distribution of log-returns for the 'Adj Close' price. (Keep in mind that in most cases, what Numpy can do on arrays, it can also do on pandas Series.)

```python
log_gsframe = np.log(gsframe[['Adj Close']])
```

```python
import matplotlib.pyplot as plt
log_return = np.diff(log_gsframe)
plt.plot(log_return)
```
